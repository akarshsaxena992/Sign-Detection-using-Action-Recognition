```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
import time
import mediapipe as mp
```

```python
#STEP 2 : Keypoints using MP Holistic
```

```python
#Set up of mediapipe holistic functions
#Mediapipe holistic is used for downloading that model and leveraging that model
mp_holistic = mp.solutions.holistic # Holistic model
#Mediapipe the drawing utilities is actually making it easier to draw key points on our face
mp_drawing = mp.solutions.drawing_utils # Drawing utilities
```

```python
# function to set and make out detections and it will be used when capturing image from webcam
# model is mediapipe holisitc model
# we first convert an image from bgr to rgb and then set it to unwritable so this saves a little bit of memory then we make out detection convert it or set it back to wriatble and then convert from rgb to bgr
# bt default the feed from opencv read the video into the format of bgr but fro detection using mediapipe we need it to be in rgb
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # COLOR CONVERSION BGR 2 RGB
    image.flags.writeable = False                  # Image is no longer writeable
    results = model.process(image)                 # Make prediction and detection using mediapipe
    image.flags.writeable = True                   # Image is now writeable 
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # COLOR COVERSION RGB 2 BGR
    return image, results
```

```python
# This function is used to visualize detection i.e. to draw landmarks on human body
def draw_landmarks(image, results):
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION) # Draw face connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS) # Draw pose connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS) # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS) # Dr
```

```python
# In this we have changed the size and color of different landmarks to make them visible
def draw_styled_landmarks(image, results):
    # Draw face connections
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION, 
                             mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1), 
                             mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
                             ) 
    # Draw pose connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                             ) 
    # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                             ) 
    # Draw right hand connections  
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                             ) 
```

```python
#cap is storing and accesssing out webcam and by saying VideoCapture(0) means we are reading from device number 0 and i.e. our webcam
#Here 0 is used fr webcam and we can also put the location of video if we wan to use video
cap = cv2.VideoCapture(0)

# Set mediapipe model and we use with statement to be able to access our holistic model 
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        #This loop is to loop through all the frames of the video from webcam
        
        # Read feed
        #This Read feed is used to return a value and a frame which is basically an image from the webcamm
        ret, frame = cap.read()

        # Make detections
        image, results = mediapipe_detection(frame, holistic)
        print(results)
        
        # Draw landmarks
        draw_styled_landmarks(image, results)
        
        #Pass the above image in this function and this is going to show up back to the user
        # Show to screen
        cv2.imshow('OpenCV Feed', image)

        # Break gracefully
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
    #This is to destry or break the webcam 
    cap.release()
    #This is used to close all the frames.
    cv2.destroyAllWindows()

```

```python
# this is same as above one, we are just converting the extracted keypoints into numpy array
def extract_keypoints(results):
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33*4)
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([pose, face, lh, rh])
```

```python
# Path for exported data, numpy arrays
DATA_PATH = os.path.join('MP_Data') 

# Actions that we try to detect
actions = np.array(['hello', 'thanks', 'iloveyou'])

# Thirty videos worth of data
no_sequences = 30

# Videos are going to be 30 frames in length
sequence_length = 30

# Folder start
start_folder = 30
```

```python
# Here we are storing each of the 30 frames of above actions inside a folder
# Looping through actions
for action in actions: 
#     Looping through 30 sequences
    for sequence in range(no_sequences): 
        try: 
            os.makedirs(os.path.join(DATA_PATH, action, str(sequence)))
        except:
            pass
```

```python
# STEP 5 : Collecting keypoint Values for training and testing
```

```python
cap = cv2.VideoCapture(0)
# Set mediapipe model 
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    # NEW LOOP
    # Loop through actions
    for action in actions:
        # Loop through sequences aka videos
        for sequence in range(no_sequences):
            # Loop through video length aka sequence length
            for frame_num in range(sequence_length):

                # Read feed
                ret, frame = cap.read()

                # Make detections
                image, results = mediapipe_detection(frame, holistic)
#                 print(results)

                # Draw landmarks
                draw_styled_landmarks(image, results)
                
                # NEW Apply wait logic
                if frame_num == 0: 
                    cv2.putText(image, 'STARTING COLLECTION', (120,200), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255, 0), 4, cv2.LINE_AA)
                    cv2.putText(image, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12), 
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
                    # Show to screen
                    cv2.imshow('OpenCV Feed', image)
                    cv2.waitKey(2000)
                else: 
                    cv2.putText(image, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12), 
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
                    # Show to screen
                    cv2.imshow('OpenCV Feed', image)
                
                # NEW Export keypoints
                keypoints = extract_keypoints(results)
                npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_num))
                np.save(npy_path, keypoints)

                # Break gracefully
                if cv2.waitKey(10) & 0xFF == ord('q'):
                    break
                    
    cap.release()
    cv2.destroyAllWindows()
```

```python
# STEP 6 : Preprocess Data and Create Labels and Features
```

```python
# This function is used to partition our data into testing and training
from sklearn.model_selection import train_test_split
# The below function is used to convert the data into one hot encoded data
from tensorflow.keras.utils import to_categorical
```

```python
# Creation of label dictionary
label_map = {label:num for num, label in enumerate(actions)}
```

```python
label_map
```

```python
# Here sequences are going to effectively represent our feature data or x data and labels are going to represent our label or y data
# Here in this block we are going through every frames and appending it to window and then appending every window to sequence which is total 90
sequences, labels = [], []
for action in actions:
    for sequence in range(no_sequences):
        window = []
        for frame_num in range(sequence_length):
            res = np.load(os.path.join(DATA_PATH, action, str(sequence), "{}.npy".format(frame_num)))
            window.append(res)
        sequences.append(window)
        labels.append(label_map[action])
```

```python
np.array(sequences).shape
```

```python
np.array(labels).shape
```

```python
X = np.array(sequences)
```

```python
X.shape
```

```python
# Here we have converted our initial labels which is going to be a series of numbers that we have passed through label map and converted it into one hot encoded label which is like a binary flag
y = to_categorical(labels).astype(int)
```

```python
y
# [1, 0, 0] represents hello 
# [0, 1, 0] represents thankyou
# [0, 0, 1] represents iloveyou
```

```python
# Here we are doing partition for training and testing of data
# here test_size indicates that testing data is 5 percent
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05)
```

```python
# STEP 7 : BUILD AND TRAIN LSTM NEURAL NETWORK
```

```python
# This is allow us to build sequential neural network
from tensorflow.keras.models import Sequential 
# This is going to give us the lstm layer and dense layer
from tensorflow.keras.layers import LSTM, Dense
# This is going to allow us to perform some logging inside of tensorboard if we wanted to go and trace and monitor our model as its training
from tensorflow.keras.callbacks import TensorBoard
```

```python
# Creating a log directory and set up our tensorboard callbacks
# Tensorboard is used to monitor testing and training
log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
# Initializing sequential model
model = Sequential()
# Adding lstm layers
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(30,1662)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
# Here softmax return values of probability within zero to one
model.add(Dense(actions.shape[0], activation='softmax'))
```

```python
# This is the result that we got after doing all the above tasks
res = [.7, 0.2, 0.1]
```

```python
# The result is going to preprocess and extract the actions
actions[np.argmax(res)]
```

```python
# compilatio of model and the optimizer that we have used is atom and then we have unsepcified loss
# loss = categorical_crossentropy so this is the loss function that you need to use when you have a multi-class classification
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
# fit and train our model 
model.fit(X_train, y_train, epochs=2000, callbacks=[tb_callback])
```
